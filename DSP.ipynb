```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.linspace(2, 3.5,500)
sample = 30
bit = 4
parts = 2**bit
```

```python
def f(x):
    return np.sin(x**2)+9*x**-2 + 5
```

```python
y = f(x)
plt.plot(x, y, "b")
plt.show()
```

```python
x_parts = np.linspace(2, 3.5, sample)
y_parts = f(x_parts)
plt.plot(x, y, "b")
plt.scatter(x_parts, y_parts, color = "blue", s = 10)
plt.show()
```

```python
#curve
plt.plot(x, y, "b")
#staircase
plt.plot([x_parts[0], x_parts[1]], [y_parts[0], y_parts[0]], "b")
for i in range(sample - 1):
    plt.plot([x_parts[i+1], x_parts[i+1]], [y_parts[i], y_parts[i+1]], "b")
    if i < sample -2:
        plt.plot([x_parts[i+1], x_parts[i+2]], [y_parts[i+1], y_parts[i+1]], "b")
plt.show()
```

```python
y_max = np.max(y_parts)
y_min = np.min(y_parts)
chunks = np.linspace(y_min, y_max, parts + 1)
chunks
```

```python
#line
plt.plot([1.9, 1.9], [y_min, y_max])
corr = (chunks[0] + chunks[1])/2 - chunks[0]
#chunks and level
for itr in range(0,parts + 1):
    plt.plot([1.8, 3.5], [chunks[itr], chunks[itr]], "b--", linewidth = 0.5)
    if itr < parts:
        plt.text(1.79, chunks[itr]+corr, itr)
#plt.plot(x, y, color = "blue")

#staircase
plt.plot([x_parts[0], x_parts[1]], [y_parts[0], y_parts[0]], "b")
for i in range(sample - 1):
    plt.plot([x_parts[i+1], x_parts[i+1]], [y_parts[i], y_parts[i+1]], "b")
    if i < sample - 2:
        plt.plot([x_parts[i+1], x_parts[i+2]], [y_parts[i+1], y_parts[i+1]], "b")
plt.show()
```

```python
#line
plt.plot([1.9, 1.9], [y_min, y_max])
corr = (chunks[0] + chunks[1])/2 - chunks[0]
#chunks and Level
for itr in range(0,parts + 1):
    plt.plot([1.8, 3.5], [chunks[itr], chunks[itr]], "b--", linewidth = 0.5)
    if itr < parts:
        plt.text(1.79, chunks[itr]+corr, itr)

#points
plt.scatter(x_parts, y_parts, color = "blue", s = 10)
plt.show()
```
