$$I_{o}=\frac{V_{o}}{\sqrt{R^2+(X_{C}-X_{L})^2}}$$
As we know,
$$X_{C} = \frac{1}{\omega C}, X_{L} = \omega L$$
And
$$\omega = 2\pi f$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def X_C(f, C):
    w = 2*np.pi*f
    xc = 1/(w*C)
    return xc
```

```python
def X_L(f, L):
    return 2*np.pi*f*L
```

```python
def I_0(V0, R, xc, xl):
    Z = np.sqrt(R**2 + (xc-xl)**2)
    return V0/Z
```

```python
V0 = 3.2
R = 100
C = 0.01 * 10**(-6)
L = 50 * 10**(-3)
f = np.linspace(6000, 8000, 10000)
I = I_0(V0, R, X_C(f, C), X_L(f, L))
plt.plot(f, I)
plt.show()
```

```python
I_max = np.max(I)
I_1 = 0.707*I_max
indx = np.where(I == I_max)
f[indx][0], I_max
```
