### Definition of Orthonormal Basis

An orthonormal basis of a vector space $ùëâ$ with an inner product (such as the dot product in $ùëÖ^ùëõ$) is a basis consisting of vectors that are:

<ol>
  <li> <strong>Orthogonal:</strong> Each pair of different basis vectors is orthogonal, meaning that their inner product is zero.</li>
  <li><strong>Normalized:</strong> Each basis vector has a norm (or length) of one.</li>
</ol>

### Mathematical Expression
<p>
If $\{e_1, e_2, ....., e_n\}$ is an orthonormal basis for $V$, then for $i, j$ among $\{1, 2, ....., n\}$,
$\langle e_i, e_j \rangle = \delta_{ij}$
where $\delta_{ij}$ is the Kronecker delta, defined as: 
$$
\delta_{ij} = 
\begin{cases} 
1 & \text{if } i = j \\
0 & \text{if } i \neq j
\end{cases}
$$
</p>

### Properties of an Orthonormal Basis
<ol>
    <li><strong>Simplification of Calculations:</strong> Computations involving vectors in terms of their components become simpler because the inner product of any two different basis vectors is zero and the square of the norm of any basis vector is one.</li>
    <li><strong>Representation of Vectors:</strong>  Any vector $v$ in $V$ can be uniquely expressed as a linear combination of the basis vectors:
    $$
        v = \sum_{i=1}^n c_ie_i
    $$
where $c_i$ are coefficients that need to be determined.</li>
    <li><strong>Preservation under Linear Transformation: </strong> In the contex of matrices , if the columns of a square matrix $A$ from an orthonormal set, then $A$ is an orthogonal matrix. Which implies that $A^TA=AA^T=I$, where $I$ is the identity matrix. this preperty is useful for various applications including solving linear equation and matrix factorizations.</li>
    <li><strong>Stability and Numerical Properties:</strong> Orthonormal bases are particularly favored in numerical linear algebra because operations involving these bases are more stable and less prone to rounding errors.</li>
     <li><strong>Application in Fourier Series and Transformations:</strong> In functional analysis, orthonormal bases are crucial for expanding functions in series (like Fourier series) where functions are represented as infinite sums of orthonormal basis functions.</li>
</ol>

### Theory
The normalized wavefunctions of the Schrodinger wave equation for a particle is:
$$ u_{n}(x) = \sqrt{\frac{2}{L}} \sin \left( \frac{n\pi x}{L} \right) ------------- (1) \; \text{for} \; 0 \le x \le L $$
Otherwise equal to 0. Values of $n$ are positive integers $1, 2, 3, ‚Ä¶‚Ä¶ $The energy eigenvalues are:
$$ E_{n} = \frac{\hbar^2 n^2 \pi^2}{2 m L^2} $$
The scalar product of two eigenfunctions is defined as:
$$ \langle u_{n}(x), u_{m}(x) \rangle = \int_{0}^L u_{n}^*(x) u_{m}(x) \, dx = \delta_{nm} ------------- (2) $$
The closure relation is :
$$\sum_n u_{n}^*(x) u_{n}(x') = \delta(x - x') ------------- (3) $$
where $\delta(x - x')$ is the Dirac delta function and the sum runs upto n equals to infinity. 

Basic Properties of Dirac Delta Function are as follows:
<ol>
    <li>$\delta(x - x') = \infty \; when \; x = x',$</li>
    <li>$\delta(x - x') = 0 \; when \; x \neq x',$</li>
    <li>$\int_{-\infty}^\infty \delta(x - x') \, dx = 1 $</li>
</ol>
Third property of Dirac delta function can be simplified as:

$1 = \int_{-\infty}^\infty \delta(x - x') \, dx = \int_{0}^L \sum_n u_{n}^*(x) u_{n}(x') \, dx = \sum_n \frac{4}{(2n - 1)\pi} \sin \left( \frac{(2n - 1)\pi x'}{L} \right) ------------- (4)$ where n can take positive integer values $1, 2, 3, ....$

The sum on the left hand side term of equation (3) for finite n can be defined as a generator of Dirac delta function as:
$$ F(x, x', n) = \sum_n u_{n}^*(x) u_{n}(x') ------------- (5) $$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
n = np.linspace(1, 100, 100)
x1 = 0.34
```

```python
def func1(x, x0):
    f = 2*np.sin(x*n*np.pi)*np.sin(x0*n*np.pi)
    return np.cumsum(f)
```

The height variation of the generator function on the value of $n$ at $x = x‚Äô = 0.114$ and $L = 1$ is shown in Figure 1. Figure 1 
shows that the height of the generator function almost linearly increases on increasing $n$ with average slope 1 and period $n 
= 10$.

```python
#plt.title('Generator of Dirac delta function as a function of n for x = x‚Äô= 0.114 and L = 1')
plt.figure(figsize=(12, 6))
plt.xlabel('n', fontsize = 18)
plt.ylabel(rf"$F(x={x1}, x'={x1})$", fontsize = 18)
y = func1(x1, x1)
plt.plot(n, y, linewidth = 2.5)
plt.grid(True)
#plt.suptitle('Figure 1: Generator of Dirac delta function as a function of n for x = x‚Äô= 0.22 and L = 1', y=0.02) 
plt.savefig('Figure 2.png', dpi=600, bbox_inches='tight')
plt.show()
```

```python
N = np.linspace(1, 1000, 1000)
```

```python
def func3(a, b, x_0):
    r =  2 / (np.pi * N) * (np.cos(np.pi * a * N) - np.cos(np.pi * b * N)) * np.sin(np.pi * N * x_0)
    return np.cumsum(r)
```

### Approximation using normal curve

The probability density for the Gaussian distribution is:
$$ p(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{- \frac{(x - \mu)^2}{2\sigma^2}} $$
where $\mu$ is the mean and $\sigma$ the standard deviation. The square of the standard deviation, $\sigma^2$, is called the variance.

The function has its peak at the mean, and its ‚Äúspread‚Äù increases with the standard deviation (the function reaches 0.607 times its maximum at $x + \sigma$ and $x - \sigma$ .This implies that normal is more likely to return samples lying close to the mean, rather than those far away. So as the mean apporaches to 0 and standard deviation is smaller the normal curve more like to behave like Dirac Delta function.

```python
def normal(x, x_0, sigma):
    exponent = (x - x_0)**2 / (2*sigma**2)
    return np.exp(-exponent)/(sigma*np.sqrt(2*np.pi))
```

```python
x = np.linspace(-2, 2, 1000)
label = [r'$\sigma = 0.2$', r'$\sigma = 0.1$', r'$\sigma = 0.01$']
ind = 0
for a in [0.2, 0.1, 0.01]:
    y = normal(x, 0, a)
    plt.plot(x, y, label = label[ind])
    plt.legend()
    ind += 1

plt.xlabel('x')
plt.ylabel(r'$p(x)$')
plt.show()
```

So, \( A(\xi) = -1 \).
Therefore, the Green's function is:
\[ G(x, \xi) =
\begin{cases}
0, & \text{if } x < \xi, \\
-e^{-\frac{\lambda}{2}(x - \xi)}, & \text{if } x \geq \xi.
\end{cases} \]
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
L = 1  # Length of the box
x_prime = 0.1  # Fixed x'
n_values = [14, 94]  # Different values of n

# Define the generator function u_n(x)
def u_n(n, x, L):
    return np.sqrt(2 / L) * np.sin(n * np.pi * x / L)

# Define the function to approximate the Dirac delta function
def dirac_delta_approx(x, x_prime, n_max, L):
    delta_approx = np.zeros_like(x)
    for n in range(1, n_max + 1):
        delta_approx += u_n(n, x, L) * u_n(n, x_prime, L)
    return delta_approx

# Generate the x values
x_values = np.linspace(0, L, 1000)

# Plot the results for different n
plt.figure(figsize=(12, 6))
for n_max in n_values:
    delta_approx_values = dirac_delta_approx(x_values, x_prime, n_max, L)
    plt.plot(x_values, delta_approx_values, label=f'n = {n_max}')

plt.title('Approximation of Dirac Delta Function')
plt.xlabel('x')
plt.ylabel(f'Sum of u_n(x) u_n({x_prime})')
plt.legend()
plt.grid(True)
plt.show()
```
